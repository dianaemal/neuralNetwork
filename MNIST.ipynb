```python
from neuralNetwork import NeuralNetwork 
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
nn = NeuralNetwork()
```

```python
csv1 = pd.read_csv("mnist_train.csv")
images1 = csv1.iloc[:, 1:].values
labels1 = csv1.iloc[:, 0].values
print(images1.shape)
print(labels1.shape)

```

```python
# show an image from training set:
img_gray = images1[1, :].reshape(28, 28)
plt.imshow(img_gray, cmap="gray")
plt.axis("off")
plt.show()
```

```python
# train 
images1 = images1 / 225.0
nn.train(images1, labels1)
```

```python
# test
csv2 = pd.read_csv("mnist_test.csv")
test_images1 = csv2.iloc[:, 1:]
test_images1 = test_images1 / 225.0
print(test_images1.shape)
test_labels1 = csv2.iloc[:, 0]
print(test_labels1.shape)
```

```python
preds = nn.predict(test_images1)
acc = nn.accuracy(test_labels1, preds)
print(acc)
```

```python
nn.draw_loss_curve()
```
